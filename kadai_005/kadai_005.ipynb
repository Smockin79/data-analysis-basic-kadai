```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# sin関数
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
axes[0, 1].plot(x, y1)

# ヒストグラム
data = np.random.normal(15, 5, 1000)
axes[0, 2].hist(data)

# 散布図
x2 = np.random.rand(50) * 10
y2 = np.random.rand(50) * 10
axes[1, 0].scatter(x2, y2)

plt.show()
```
